# Introducción al Análisis de Componentes Principales


## ¿Qué es?

- Los grandes conjuntos de datos son cada vez más comunes y, a menudo, difíciles de interpretar.
  
- Supongamos que tenemos una gran cantidad de variables observables $X_1,\dots,X_k$

- ¿Necesitamos todas?

- No necesariamente, si muchas son muy similares entre sí y están altamente correlacionados, tal vez la pregunta no es "¿cuál de ellas es importante?", sino más bien "¿cuál combinación lineal es importante?".

- Por ejemplo, si tienes un conjunto de datos con una gran cantidad de calificaciones de exámenes tomados por estudiantes, y estás tratando de predecir algún resultado (calificaciones en clases futuras, éxito en la universidad, salarios futuros), lo que podría ser el mejor predictor es un promedio, o una combinación lineal, de las calificaciones de los exámenes.

- PCA es una técnica de aprendizaje no supervisado que facilita esta tarea.

- PCA reproyecta los datos a un nuevo espacio vectorial de forma tal que retengamos la máxima información posible. 

- Como se mide información? En PCA y en otros métodos estadísticos utilizamos la varianza. 

- Si algo varía más entonces tiene más información.

- PCA identifica los ejes con la mayor  variación en los datos.

- PCA "transforma" las variables originales en un nuevo conjunto de variables, los componentes principales, que no están correlacionados y que se ordenan de manera que los primeros retengan la mayor parte de la variación presente en todas las variables originales.

- PCA se refiere entonces al proceso por el cual los componentes principales son calculados, y el uso de estos componentes para entender los datos.

## Como lo hace? 

- Reproyectando a través de combinaciones lineales

- Por ejemplo, si tenemos 7 variables ($X_1, X_2, \dots, X_{7} $), el primer componente principal va a buscar la combinación lineal de estas 7 variables de forma tal que la información retenida sea la más alta posible
    
    
\begin{align}
PC_1 = \delta_{1,1} X_1+ \delta_{2,1} X_2 + \dots + \delta_{7,1} X_{7}
\end{align}


en donde $\delta_1$ es un vector de $7$ números reales $(\delta_1= \delta_{1,1},\dots, \delta_{7,1})$, que llamamos pesos o loadings y han sido normalizados para que $\sum_{j=1}^{7} \delta_{j1}^2=1$
 

- Podemos entonces calcular los componentes principales restantes, agregando la restricción de que este no este correlacionado con PC1.



# Aplicación

## Motivación: medir algo que no vemos

A los economistas y hacedores de políticas, muchas veces nos interesa estudiar el efecto de políticas educativas sobre el aprendizaje de los estudiantes. Un ejemplo es el paper de Duflo, Dupas y Kremer (2011) ["Peer Effects, Teacher Incentives, and the Impact of Tracking: Evidence from a Randomized Evaluation in Kenya"](https://www.aeaweb.org/articles?id=10.1257/aer.101.5.1739), que evalúa el impacto de **agrupar a los estudiantes por habilidad** ("tracking") en escuelas de Kenia.



***Abstract***

*To the extent that students benefit from high-achieving peers, tracking will help strong students and hurt weak ones. However, all students may benefit if tracking allows teachers to better tailor their instruction level. Lower-achieving pupils are particularly likely to benefit from tracking when teachers have incentives to teach to the top of the distribution. We propose a simple model nesting these effects and test its implications in a randomized tracking experiment conducted with 121 primary schools in Kenya. While the direct effect of high-achieving peers is positive, tracking benefited lower-achieving pupils indirectly by allowing teachers to teach to their level. (JEL I21, J45, O15)* 

***Resumen***

*En la medida en que los estudiantes se benefician de sus compañeros con alto rendimiento, el seguimiento escolar beneficiará a los estudiantes más destacados y perjudicará a los menos destacados. Sin embargo, todos los estudiantes pueden beneficiarse si el seguimiento escolar permite a los docentes adaptar mejor su nivel de instrucción. Los alumnos con bajo rendimiento académico tienen una mayor probabilidad de beneficiarse del seguimiento escolar cuando los docentes tienen incentivos para enseñar a los alumnos con mayor rendimiento. Proponemos un modelo simple que anida estos efectos y comprobamos sus implicaciones en un experimento de seguimiento aleatorio realizado con 121 escuelas primarias en Kenia. Si bien el efecto directo de los compañeros con alto rendimiento es positivo, el seguimiento escolar benefició indirectamente a los alumnos con bajo rendimiento al permitir a los docentes enseñar a su nivel. (JEL I21, J45, O15)*




Pero esto nos plantea una pregunta fundamental:

> ¿Cómo medimos la habilidad si no la observamos directamente?

Lo que tenemos son resultados de exámenes en distintas materias. Lo que nos gustaría tener es una medida sintética de **habilidad general**, algo así como un GPA. Pero no queremos promediar arbitrariamente las notas, como lo hace el GPA tradicional. Queremos construir una variable latente no observada directamente, que capture de forma parsimoniosa la estructura común detrás de las calificaciones.

Esta es la lógica detrás de un **modelo de factor**:

* Suponemos que hay una **habilidad general no observada** que influye en el desempeño de los estudiantes en todas las materias.
* A partir de las notas, podemos tratar de inferir ese factor común.
* PCA ofrece una forma empírica de aproximar este factor latente a partir de los datos, sin imponer a priori los pesos. En lugar de asumirlos, los infiere maximizando la varianza explicada por una combinación lineal de las calificaciones.
* No es algo que medimos directamente con un termómetro.
* Lo calculamos como un **resumen de varias calificaciones** para representar una habilidad académica subyacente.

En notación:

$$
\text{habilidad}_i \quad \text{no observable directamente}
$$

pero observamos

$$
x_{i1} = \text{nota en Adiciones}, \dots , \quad x_{ij} = \text{nota en Ortografía}, \quad x_{ik} = \text{nota en Reconocimiento de Palabras}, \dots
$$



##  PCA como modelo de factor

El **modelo de factor** más simple dice:

$$
x_{ij} = \mu_j + \lambda_j F_i + \varepsilon_{ij}
$$

donde:

* $x_{ij}$ = puntaje del individuo $i$ en la materia $j$
* $\mu_j$ = media de la materia $j$
* $\lambda_j$ = carga (peso) que relaciona el factor con la materia
* $F_i$ = factor común (habilidad académica) del individuo $i$
* $\varepsilon_{ij}$ = componente específico o ruido de la materia $j$

**Interpretación:**

* Todas las materias comparten un **patrón común** (habilidad académica) que explica parte de la variación de las notas.
* El resto es ruido o habilidad específica para esa materia.



## Lo que hace PCA

PCA **no sabe** que existe "habilidad académica", pero busca **la combinación lineal de las notas que explique más varianza**.

En el modelo de factor:

* El primer componente principal estimado por PCA es

  $$
  \hat{F}_i = w_1 x_{i1} + w_2 x_{i2} + \dots + w_p x_{ip}
  $$

  donde los pesos $w_j$ son elegidos para **maximizar la varianza** de $\hat{F}_i$, sujeto a que $\|w\| = 1$.
* Estos $w_j$ estiman proporcionalmente las **cargas** $\lambda_j$ del modelo de factor.

Así, el **primer componente principal ≈ factor más importante** que explica los datos.
En el ejemplo del GPA:

* Si todas las notas están positivamente correlacionadas, el primer componente principal tendrá pesos positivos en todas, como un GPA optimizado para explicar varianza.
* El segundo componente buscaría patrones ortogonales, como diferencias entre habilidades (ej. más verbal vs. más matemática).



## GPA vs PCA

* **GPA**: promedia todas las materias con pesos iguales (1/p).
* **PCA**: busca los pesos que **maximizan la varianza explicada**, no los fija arbitrariamente.

## **Extensión (2 factores): No hay una sola habilidad**

Hasta ahora supusimos que existe un único patrón común de desempeño, una sola dimensión de habilidad. Pero esto puede ser demasiado simplificador.

En el contexto del estudio de Duflo, Dupas y Kremer, es razonable pensar que los estudiantes no poseen una única habilidad general, sino que su desempeño académico está determinado por **distintas dimensiones cognitivas**. Por ejemplo:

* **Habilidades lógico-matemáticas**, reflejadas en pruebas como adiciones.
* **Habilidades lingüísticas**, capturadas por ejercicios de ortografía o reconocimiento de palabras.

Esto implica que las calificaciones que observamos podrían estar determinadas por **más de un factor latente**.

PCA permite detectar justamente eso:

* Extrae **más de un componente principal**, cada uno representando una dimensión ortogonal que resume la estructura de correlación entre variables.
* El **primer componente** (PC1) es la combinación lineal que explica la mayor parte de la varianza común.
* El **segundo componente** (PC2), ortogonal al primero, captura la siguiente fuente importante de variabilidad.

Desde una perspectiva empírica, esto es útil porque:

* Nos permite representar a los estudiantes en un plano PC1–PC2 y explorar **diferencias en sus perfiles de habilidad**.
* Facilita análisis como el *tracking* o segmentación de estudiantes, no solo en función de un promedio general, sino considerando **perfiles cognitivos diferenciados**.
* También permite estudiar si los efectos de políticas o intervenciones educativas, como la composición del grupo o los incentivos a docentes, varían según la dimensión de habilidad afectada.

> En resumen: cuando la habilidad no es unidimensional, PCA nos da una herramienta para inferir y visualizar múltiples factores relevantes, sin necesidad de especificarlos ex ante.



### Lo que hace PCA cuando hay 2 factores

PCA **no sabe de antemano** que hay dos factores latentes, pero al analizar la **estructura de correlaciones** entre las materias puede revelar esa información.

* El **primer componente principal (PC1)** es la combinación lineal de las materias que **explica la mayor parte de la varianza total** en los datos.

  * Si existen dos grupos de materias que están fuertemente correlacionadas entre sí (por ejemplo, un grupo de materias cuantitativas y otro de verbales), PC1 tenderá a reflejar la dimensión que concentra **más varianza compartida**.
  * Esto no significa que PC1 sea una mezcla arbitraria de factores, sino que **resume el patrón dominante de covariación**.

* El **segundo componente principal (PC2)** es una combinación lineal de las materias que:

  * Es **ortogonal a PC1** (no está correlacionada con él).
  * Captura la **siguiente fuente independiente de variación** en los datos.
  * Si los grupos de materias están bien definidos, PC2 tenderá a reflejar la **segunda dimensión de habilidad** que PC1 no capturó.

En este sentido, **PCA actúa como una lupa que revela las principales direcciones de variación en los datos**, y estas direcciones pueden coincidir con factores latentes significativos, como distintos tipos de habilidades académicas.



## Visualización: *Biplot*

El **biplot** es una herramienta fundamental para interpretar visualmente los componentes principales.

* Cada **estudiante** aparece como un punto en el plano definido por los primeros dos componentes principales (PC1 y PC2).

  * Las coordenadas de cada punto corresponden a los *scores* del estudiante en estos dos componentes. 
    $$
        \text{score}_{i1} = w_1^\top x_i \quad \text{(PC1)}, \qquad \text{score}_{i2} = w_2^\top x_i \quad \text{(PC2)}, \dots
        $$
    *donde $x_i$ es el vector de notas del estudiante $i$ y $w_1, w_2$ son los vectores de pesos (loadings) del primer y segundo componente, respectivamente.*


  * Esto permite identificar patrones, agrupamientos y posibles extremos (outliers).

* Cada **flecha** representa una **asignatura** (una variable original).

  * La dirección de la flecha indica **cómo se proyecta esa variable** sobre los componentes principales.
  * Su longitud indica qué tan bien está representada por ese plano (más larga = más varianza explicada por PC1 y PC2).

* Si todas las flechas apuntan más o menos en la **misma dirección**, es señal de que hay una **dimensión común fuerte** compartida entre las materias, es decir, un **factor general de habilidad**.

* En cambio, si las flechas se agrupan en **dos direcciones distintas**, eso sugiere la presencia de **dos factores diferenciados**, como:

  * Un grupo de materias relacionadas con habilidad matemática.
  * Otro grupo relacionado con habilidad verbal.

> El biplot no solo ayuda a interpretar los componentes, sino también a **visualizar las relaciones entre estudiantes y materias** en un mismo gráfico.


```python
# Carguemos el paquete
from pca import pca

model_estudiantes = pca(n_components=2,normalize=True, verbose=0)
results_estudiantes = model_estudiantes.fit_transform(X)

```

```python
# Biplot

model_estudiantes.biplot(PC=[0, 1], alpha=0.1,legend=False, color_arrow="red", arrowdict={'color_text':"black"})
```



### Interpretación de los estudiantes en el biplot

Cada punto azul representa un estudiante, y su posición corresponde a sus **scores** en los componentes principales:

* El eje horizontal es PC1 (por ejemplo, habilidad general).
* El eje vertical es PC2 (por ejemplo, habilidad especifica).

La **distancia al origen (0,0)** y la dirección de cada punto permiten interpretaciones útiles:

* Estudiantes **cerca del centro** tienen un perfil más "típico" o promedio, tanto en desempeño general como en la diferencia entre áreas.
* Estudiantes **alejados del centro** presentan **perfiles más extremos**:

  * Si se alejan hacia la derecha o izquierda: se destacan o rezagan en la dimensión capturada por PC1.
  * Si se alejan hacia arriba o abajo: lo hacen en la dimensión capturada por PC2.


> Así, el biplot no solo resume la información multivariada, sino que permite identificar **perfiles académicos diferenciados**, extremos o balanceados, dentro del grupo de estudiantes.


```python
scores = results_estudiantes['PC']
scores
```

```python
sorted_scores_pc1 = scores.sort_values(by='PC1', ascending=False)
sorted_scores_pc1
```

```python
# Plot the density estimation of the sorted scores for PC1
plt.figure(figsize=(10, 6))
sb.kdeplot(sorted_scores_pc1['PC1'], fill=True)
plt.title('Density Estimation of Sorted Scores for PC1')
plt.xlabel('PC1 Scores')
plt.ylabel('Density')
plt.show()
```
